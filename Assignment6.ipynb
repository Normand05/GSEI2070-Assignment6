```python
ds = xr.open_dataset('/glade/work/kumar34/GSEI2070_DAT/cru_ts4.09.1901.2024.tmp.dat.nc')
ds
```

```python
print("Dimensions:")
print(ds.dims)

print("\nVariables:")
print(ds.data_vars)

print("\nCoordinates:")
print(ds.coords)

print("\nAttributes:")
print(ds.attrs)

print("\nNumber of time steps:")
print(len(ds.time))
```

```python
lat_res = ds.lat.values[1] - ds.lat.values[0]
lon_res = ds.lon.values[1] - ds.lon.values[0]

print("Latitude resolution:", lat_res)
print("Longitude resolution:", lon_res)
print("Start time:", ds.time.min().values)
print("End time:", ds.time.max().values)
```

```python
#The dataset has dimensions of 1488 time steps, 360 latitude points, and 720 longitude points. The main variable stored is temperature (tmp). The spatial resolution is 0.5° × 0.5°. The dataset was produced by the British Atmospheric Data Centre (RAL, UK). The temporal resolution is monthly, and the dataset spans 1488 months from 1901 to 2024.
```
